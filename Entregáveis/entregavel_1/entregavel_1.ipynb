## Setup

Importa as principais bibliotecas e dataframe que serão utilizadas para análise 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "J:\Faculdade\Turing USP\TA\Entregáveis\entregavel_1\dataset1.csv" #caminho do arquivo no meu PC, substituir pelo seu
df = pd.read_csv(path) #dataframe
```

```python
df.head() 
```

## Item A:

```python
df.info()
```

```python
semana = df['data'].str[:-19] #Nova coluna dia da semana
df['semana'] = semana

df['data'] = df['data'].str[-16:] #Remove dia da semana de data

hora = df['data'].str[11:13] #Nova coluna hora
df['hora'] = hora

dia = df['data'].str[:2] #Nova coluna dia do mês
df['dia'] = dia

df.head()
```

```python
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y %H:%M') #Data Object -> Datetime64 
df.head()
```

```python
df.info()
```

## Item B:

```python
sns.lineplot(data=df, x="hora", y="temperatura", color='#CC0876')
plt.ylabel("Temperatura", size=10)
plt.xlabel("Tempo", size=10)
plt.title("Gráfico de Temperatura em função do tempo", size=12)
plt.show()
```

É possível perceber que em média as temperaturas se encontram mais baixas durante as primeiras horas do dia e se elevam até 15/16 horas onde apresentam seu pico e começam a decair até as primeiras horas do dia seguinte, ou seja, com o ínicio da manhã temos um aumento de temperatura que permanece até o início do anoitecer. .

## Item F:

```python
temperatura_max = df.groupby(['dia'])['temperatura'].max() #Armazena a temperatura máxima de cada dia
temperatura_min = df.groupby(['dia'])['temperatura'].min() #Armazena a temperatura mínima de cada dia
media_max = temperatura_max.mean() #Armazena a a média da temperatura mínima mensal
media_min = temperatura_min.mean() #Armazena a a média da temperatura máxima mensal
```

```python
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(temperatura_max, label='max', color = 'lightcoral')
plt.plot(temperatura_min, label='min', color = 'aqua')
plt.axhline(media_max, color = 'red', linestyle='dashed', label = 'mean max')
plt.axhline(media_min, color = 'blue', linestyle='dashed', label = 'mean min')
plt.legend()
plt.ylabel("Temperatura (ºC)", size=10)
plt.xlabel("Tempo", size=10)
plt.title("Gráfico de Temperatura em função do tempo", size=12)
plt.show()
```
